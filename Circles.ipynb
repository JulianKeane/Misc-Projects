```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Constants
DEGREES = np.linspace(0,2*np.pi,100)
```

```python
'''
Method that handles pretty much all the logic. See plotCircle for more details
'''
def Circle(r, dim, origin = [0,0], fill = False):
    x = []
    y = []
    if(len(origin) != 2):
        print("Invalid origin; using default")
        origin = [0,0]
    for degree in DEGREES:
        x.append(r*np.cos(degree)+origin[0])
        y.append(r*np.sin(degree)+origin[1])
    if(fill is True):
        fillCircle(r, origin, dim)
    plt.plot(x,y, '-b')
    
    
'''
Method that 'fills in' the circle by creating many circles with radius 0 to r; number of circles is dependent on
the size of your figure
'''
def fillCircle(r, origin, dim):
    circs = np.linspace(0,r,dim**2*4+18)
    for circ in circs:
        Circle(circ, origin)
        
        
'''
plots a Circle onto a plot. 

r is a int or float that specificies the radius of the circle
origin is a list that specificies where the center of the circle is in Cartesian coordinates
fill is a boolean that determines whether the circle is filled or an outline
dim is the dimension of the window
'''
def plotCircle(r, origin = [0,0], fill = False, dim = 4):
    plt.figure(figsize = (dim,dim))
    Circle(r = r, origin = origin, fill = fill, dim = dim)
    plt.show()
```

```python
plotCircle(1, dim = 2, fill = True)
```

```python
plotCircle(6, fill = True)
```

```python
plotCircle(1.54, origin = [2,6], dim = 3)
```
